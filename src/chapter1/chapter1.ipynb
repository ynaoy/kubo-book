<h1>1.1 なぜ「統計」な「モデル」？</>

統計モデルは、<br>
・観測によってデータ化された現象を説明するために作られる<br>
・確率分布が基本的な部品であり、これはデータにみられるばらつきを表現する手段である<br>
・データとモデルを対応付ける手続きが準備されていて、モデルがデータにどれぐらいよく当てはまっているかを定量的に評価できる


現象の観測や実験で得られたデータに対して、人間の解釈をあてはめて、そのデータの背後にある「しくみ」を理解することが目的。<br>
このときに二段階の情報消失が発生する

第一段階目では現象を「観測データ」に変換する。<br>
現象が持つ情報はあまりにも大きく、そのすべてを人間が観測・測定することは不可能

第二段階目では統計モデルを使ったデータ解析によってさらに情報が失われる<br>

データ化された自然ですら人間にとっては複雑なので要約・整理する必要がある<br>
統計モデルのあてはめによって情報が整理されて、<br>
「こういう要因でこう変化する」といった人間に理解できる部分と<br>
「ノイズ」的に見える部分に分解できる


<h1>1.2「ブラックボックスな統計解析」の悪夢</>

データ解析で使っている統計モデルをあまり理解できていない、そこに統計モデルがあることに気づいていないことがある。<br>
統計ソフトウェアにデータを入れればそれらしい出力が得られるので何も問題はないように思ってしまう。<br>
・「有意差」がでるまで検定手法をひたすら取り換える<br>
・観測者同士の割り算によって、新しい「指標をでっちあげる」<br>
・データを示すときには何でも検定してＰ値をつける、Ｐ値が小さいほど主張は正しい<br>
など

このようなブラックボックスな統計学を避け、解析者がデータをよく見て、解析の目的に沿いつつその構造に合致した統計モデルを構築することが目的。
